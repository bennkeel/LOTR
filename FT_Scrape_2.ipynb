```python
import requests
from bs4 import BeautifulSoup
import numpy as np

# URL of the web page to scrape
url = 'https://scryfall.com/card/ltr/1/banish-from-edoras'  # Replace with the actual URL

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all <div> tags with class "card-text-flavor"
div_tags = soup.find_all('div', class_='card-text-flavor')

# Extract the <p> text from each <div> tag
for div_tag in div_tags:
    
    p_tags = div_tag.find_all('p')
    for p_tag in p_tags:
        print(p_tag.get_text())
```

```python
import pandas as pd

#To get all the pages with the flavor texts
url_base = "https://scryfall.com/sets/ltc"

# Send a GET request to the URL
response_all = requests.get(url_base)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response_all.content, 'html.parser')

#find all <img> with the border-black class
img_tags = soup.find_all('a', class_='card-grid-item-card')

flavorText = []
color = []

# Extract the <p> text from each <div> tag
for link in img_tags:
    urlSub = link.get('href')
    responseSub = requests.get(urlSub)
    soupSub = BeautifulSoup(responseSub.content, 'html.parser')
    # Find all <div> tags with class "card-text-flavor"
    div_tags = soupSub.find('div', class_='card-text-flavor')
    # Extract the <p> text from each <div> tag
    if (div_tags):
        p_tag = div_tags.find('p')
        flavorText.append(p_tag.get_text())
        print(p_tag.get_text())
    else: 
        flavorText.append("NA")
        print("NA")
    #Color tags from the title
    title = soupSub.find_all('h1', class_='card-text-title')
    for thing in title:
        abbr_tags = thing.find_all('abbr', class_="card-symbol")
        cost = ""
        for abbr_tag in abbr_tags:
            cost += abbr_tag.get_text()
        print(cost)
        color.append(cost)

# Create a DataFrame from the extracted data
df = pd.DataFrame(flavorText, columns=['Text'])
```

```python
cardNames = []

for link in img_tags:
    cardName = link.find('span', class_="card-grid-item-invisible-label")
    cardNames.append(cardName.get_text())
    print(cardName.get_text())
```

```python
colors = ["W", "U", "B", "R", "G", "C", "M"]

cardColorList = []

#Check if df_all[cost] contains one or more elements from colors. Append the value to cardColorList. 
for cardCost in df_all['cost']:
    matches = [symbol for symbol in colors if symbol in cardCost]
   
    if len(matches) == 1:
        cardColorList.append(matches[0])
    elif len(matches) > 1:
        cardColorList.append('M')
    else:
        cardColorList.append('C')

# Add cardColorList as a new column to the DataFrame
df['cardColorList'] = cardColorList

# Print the modified DataFrame
print(df)
```

```python
df2 = pd.DataFrame(cardNames, columns=['card_name'])
df3 = pd.DataFrame(color, columns=['cost'])
df_all = pd.concat([df, df2, df3], axis=1)
display(df_all)
```

```python
# Export the DataFrame to a CSV file
df_all.to_csv('scryfall_extract.csv', index=False, encoding="utf-8")
```
